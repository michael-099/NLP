````python
from nltk.tokenize import word_tokenize
import re
import nltk
nltk.download('punkt_tab')

def tokenize_amharic_nltk(text):
    # space out punctuation to separate the word from the punctiation
    text = re.sub(r'([።፧፡!?;,.)()፣])', r' ', text)
    tokens = word_tokenize(text)
    tokens = [t.strip(' "\'') for t in tokens]

    return tokens

tokens = tokenize_amharic_nltk(text[:100000])
print(f"Total tokens: {len(tokens):,}")
print("tokens:", tokens)
````

### Interpolated Trigram Language Model with Laplace Smoothing

We compute the probability of a sentence $ w_1, w_2, \dots, w_n $ using **linear interpolation** of trigram, bigram, and unigram probabilities (also known as **deleted interpolation** or **Jelinek-Mercer smoothing**):

$$
P(w_1, \dots, w_n) = \prod_{i=3}^{n} P(w_i \mid w_{i-2}, w_{i-1})
$$

with padding: $ w_{-1} = w_0 = \langle s \rangle $, and final token $ \langle /\!s \rangle $.

#### Core Probability: Linear Interpolation
 combines the probabilities from unigram, bigram, and trigram models (or more) using fixed weights. instead of trusting any one model alone,blend them.

$$
P(w_i \mid w_{i-2}, w_{i-1}) =
\lambda_3 \cdot P_3(w_i \mid w_{i-2}, w_{i-1}) +
\lambda_2 \cdot P_2(w_i \mid w_{i-1}) +
\lambda_1 \cdot P_1(w_i)
$$

where $ \lambda_1 + \lambda_2 + \lambda_3 = 1 $  
(Here: $ \lambda_1 = 0.1 $, $ \lambda_2 = 0.3 $, $ \lambda_3 = 0.6 $)

#### Smoothed Component Probabilities (Add-1 / Laplace)

$$
P_3(w_i \mid w_{i-2}, w_{i-1}) =
\begin{cases}
\dfrac{C(w_{i-2},w_{i-1},w_i) + 1}{C(w_{i-2},w_{i-1}) + V} & \text{if } C(w_{i-2},w_{i-1}) > 0 \\
P_2(w_i \mid w_{i-1}) & \text{otherwise (backoff)}
\end{cases}
$$

$$
P_2(w_i \mid w_{i-1}) =
\begin{cases}
\dfrac{C(w_{i-1},w_i) + 1}{C(w_{i-1}) + V} & \text{if } C(w_{i-1}) > 0 \\
P_1(w_i) & \text{otherwise}
\end{cases}
$$

$$
P_1(w_i) = \dfrac{C(w_i) + 1}{N + V}
$$

where:
- $ C(\cdot) $ = count in training corpus
- $ V $ = vocabulary size.
   Because you added 1 for every word in the vocabulary.redistributing probability mass across all possible next words
- $ N $ = total number of word tokens in corpus

#### Key Advantages of This Model

- **Never zero probability** (thanks to Laplace smoothing + interpolation)
- **Smooth fallback**: trigram → bigram → unigram
- **Robust to unseen trigrams/bigrams** (very common in low-resource languages like Amharic)
- **Monotonic and interpretable** behavior
- Standard in academic and real-world n-gram models

````python
# 1.6 What is the probability of the sentence. "ኢትዮጵያ ታሪካዊ ሀገር ናት ". You can also try more sentences.
import math
from collections import defaultdict

# Laplace smoothing
alpha = 1

# Lambda weights for interpolation
lambda_1, lambda_2, lambda_3 = 0.1, 0.3, 0.6

# Vocabulary size
vocab_size = len(unigram_freq)


def interpolated_trigram_prob(w1, w2, w3):
    """Return interpolated probability P(w3 | w1, w2)."""

    #  1. Unigram probability
    total_unigrams = sum(unigram_freq.values())
    p1 = (unigram_freq.get(w3, 0) + alpha) / (total_unigrams + alpha * vocab_size)

    #  2. Bigram probability
    bigram_context_count = unigram_freq.get(w2, 0)
    p2 = (bigram_freq.get((w2, w3), 0) + alpha) / (bigram_context_count + alpha * vocab_size)

    #  3. Trigram probability
    trigram_context_count = bigram_freq.get((w1, w2), 0)
    p3 = (trigram_freq.get((w1, w2, w3), 0) + alpha) / (trigram_context_count + alpha * vocab_size)

    #  Linear Interpolation
    return (lambda_3 * p3) + (lambda_2 * p2) + (lambda_1 * p1)


def sentence_probability(sentence, n=3):
    """Return full sentence probability using interpolated trigram model."""

    words = tokenize_amharic_nltk(sentence.lower())

    # Add start and end markers
    words = ["<s>"] * (n - 1) + words + ["</s>"]

    prob = 1.0

    for i in range(n - 1, len(words)):
        w1, w2, w3 = words[i - 2], words[i - 1], words[i]
        prob *= interpolated_trigram_prob(w1, w2, w3)

    return prob


# TESTING

sent1 = "ኢትዮጵያ ታሪካዊ ሀገር ናት"
print(f"Sentence: {sent1}")
print("Probability ≈", sentence_probability(sent1))

test_sentences = [
    "አዲስ አበባ የኢትዮጵያ ዋና ከተማ ነች",
    "ዛሬ ደስ ይላል",
    "ፀሐይ በምትወጣበት ጊዜ ወፎች ይዘምራሉ"
]

for s in test_sentences:
    print(f"\n{s} → {sentence_probability(s)}")
````

### Random Text Generation with N-gram Models

We generate sentences by **sampling from the empirical n-gram distribution** learned from the corpus (maximum likelihood decoding via weighted random choice).

#### Bigram Generation (n=2)

Start with a random word $ w_1 $ (biased toward frequent unigrams):

$$
w_1 \sim P(w) = \frac{C(w)}{N}
$$

Then iteratively sample the next word conditioned only on the previous one:

$$
w_{t+1} \sim P(w_{t+1} \mid w_t) = \frac{C(w_t, w_{t+1})}{C(w_t)}
$$

Stop on end-of-sentence punctuation (።, ፧, !) or max length.

Result: locally correct word transitions, but often repetitive and topic-drift.

#### Trigram Generation (n=3)

Start with a frequent bigram $ (w_1, w_2) $ drawn from the most common bigrams:

$$
(w_1, w_2) \sim \text{top-k bigrams by count}
$$

Then sample each subsequent word conditioned on the previous two:

$$
w_{t} \sim P(w_t \mid w_{t-2}, w_{t-1}) = \frac{C(w_{t-2}, w_{t-1}, w_t)}{C(w_{t-2}, w_{t-1})}
$$

Stop on punctuation or max length.

Result: significantly more coherent syntax, natural phrasing, and longer grammatical units.

#### Effect of N-gram Order (Empirical Observation)

| Order | Coherence | Typical Behavior |
|-------|-----------|------------------|
| n=1 (unigram)   | Very low     | Random word salad |
| n=2 (bigram)    | Moderate     | Locally correct, but repetitive and jumps topics |
| n=3 (trigram)   | High         | Fluent, natural-sounding Amharic sentences |
| n≥4             | Very high    | Excellent fluency, but risk of reproducing exact training sentences (overfitting) |

This ancestral sampling method is the classic way neural language models (like GPT) were preceded — and remains one of the most interpretable and controllable generation techniques, especially powerful in low-resource languages like Amharic.

````python
# 1.7 Generate random sentences using n-grams; explain what happens as n-increases based on your output.
import random

#   HELPERS

END_TOKENS = {"።", "፧", "!"}

def weighted_choice(pairs, freq_dict):
    """Choose an item from list of keys using their frequencies."""
    return random.choices(
        population=pairs,
        weights=[freq_dict[p] for p in pairs],
        k=1
    )[0]

#   UNIGRAM GENERATOR

def generate_sentence_unigram(max_len=12):
    words = []

    vocab = [
        w if isinstance(w, str) else w[0]
        for w in unigram_freq.keys()
    ]

    freq = {
        (w if isinstance(w, str) else w[0]): c
        for w, c in unigram_freq.items()
    }

    for _ in range(max_len):
        w = weighted_choice(vocab, freq)
        words.append(w)
        if w in END_TOKENS:
            break

    return " ".join(words)


#   BIGRAM GENERATOR

def generate_sentence_bigram(start_word=None, max_len=15):

    def _flatten(w):
        return w if isinstance(w, str) else w[0]

    # flatten keys for start-word selection
    flat_bigrams = { ( _flatten(a), _flatten(b) ): c
                     for (a,b), c in bigram_freq.items() }

    if start_word is None:
        start_word = random.choice([a for (a,_) in flat_bigrams.keys()])

    sentence = [start_word]

    for _ in range(max_len):
        last = sentence[-1]

        candidates = [(a,b) for (a,b) in flat_bigrams if a == last]
        if not candidates:
            break

        next_word = weighted_choice(
            [b for (_,b) in candidates],
            {b: flat_bigrams[(last,b)] for (_,b) in candidates}
        )

        sentence.append(next_word)

        if next_word in END_TOKENS:
            break

    return " ".join(sentence)


#   TRIGRAM GENERATOR

def generate_sentence_trigram(max_len=15):

    def _flatten(w):
        return w if isinstance(w, str) else w[0]

    flat_trigrams = {
        (_flatten(a), _flatten(b), _flatten(c)): count
        for (a,b,c), count in trigram_freq.items()
    }

    # pick a common starting bigram
    start_bigram = random.choice(list(flat_trigrams.keys()))[:2]
    sentence = [start_bigram[0], start_bigram[1]]

    for _ in range(max_len):
        w1, w2 = sentence[-2], sentence[-1]

        candidates = [(a,b,c) for (a,b,c) in flat_trigrams if a == w1 and b == w2]
        if not candidates:
            break

        next_word = weighted_choice(
            [c for (_,_,c) in candidates],
            {c: flat_trigrams[(w1,w2,c)] for (_,_,c) in candidates}
        )

        sentence.append(next_word)

        if next_word in END_TOKENS:
            break

    return " ".join(sentence)


#   HYBRID BACKOFF MODEL

def generate_sentence_hybrid(max_len=20):
    """
    Trigram → Bigram → Unigram backoff generator.
    Works with tuple-based n-gram keys.
    """
    def _flatten(w):
        return w if isinstance(w, str) else w[0]

    # Flatten n-grams for safe access
    flat_unigram = {_flatten(w): c for w, c in unigram_freq.items()}
    flat_bigram  = {(_flatten(a), _flatten(b)): c for (a,b), c in bigram_freq.items()}
    flat_trigram = {(_flatten(a), _flatten(b), _flatten(c)): c for (a,b,c), c in trigram_freq.items()}

    # --- STARTING BIGRAM ---
    start_bigram = random.choice(list(flat_bigram.keys()))
    sentence = [start_bigram[0], start_bigram[1]]

    for _ in range(max_len):

        w1, w2 = sentence[-2], sentence[-1]

        # --- TRIGRAM ---
        trigram_candidates = [(a,b,c) for (a,b,c) in flat_trigram if a==w1 and b==w2]
        if trigram_candidates:
            next_word = weighted_choice(
                [c for (_,_,c) in trigram_candidates],
                {c: flat_trigram[(w1,w2,c)] for (_,_,c) in trigram_candidates}
            )
            sentence.append(next_word)
            if next_word in END_TOKENS:
                break
            continue

        # --- BIGRAM BACKOFF ---
        bigram_candidates = [(a,b) for (a,b) in flat_bigram if a==w2]
        if bigram_candidates:
            next_word = weighted_choice(
                [b for (_,b) in bigram_candidates],
                {b: flat_bigram[(w2,b)] for (_,b) in bigram_candidates}
            )
            sentence.append(next_word)
            if next_word in END_TOKENS:
                break
            continue

        # --- UNIGRAM BACKOFF ---
        next_word = weighted_choice(list(flat_unigram.keys()), flat_unigram)
        sentence.append(next_word)
        if next_word in END_TOKENS:
            break

    return " ".join(sentence)

#   DEMO

print(" UNIGRAM")
for _ in range(3):
    print(generate_sentence_unigram())

print("\n BIGRAM")
for _ in range(3):
    print(generate_sentence_bigram())

print("\n TRIGRAM")
for _ in range(3):
    print(generate_sentence_trigram())

# print("\n HYBRID ")
# for _ in range(5):
#     print(generate_sentence_hybrid())
````



This part of the implementation constructs a **smoothed interpolated trigram language model** on a large Amharic text corpus and performs **both intrinsic (perplexity) and extrinsic (text generation quality) evaluation**.

#### 1. **N-gram Extraction**
The function `generate_ngrams(tokens, n)` extracts all contiguous sequences of `n` tokens from the tokenized text:

```python
ngram_i = (w_i, w_{i+1}, ..., w_{i+n-1})  for i ∈ [0, ..., L−n]
```

- **Unigrams**: `n=1` → `(w_i)`
- **Bigrams**: `n=2` → `(w_{i−1}, w_i)`
- **Trigrams**: `n=3` → `(w_{i−2}, w_{i−1}, w_i)`

This yields clean, flat tuples of strings (no nested structure), enabling efficient counting and lookup.

#### 2. **Train/Test Split**
The corpus is split chronologically:
- **Training set**: First 90% of tokens → used to estimate probabilities
- **Test set**: Last 10% → reserved for intrinsic evaluation (never seen during training)

This simulates real-world deployment and prevents data leakage.

#### 3. **Frequency Counting (Maximum Likelihood Estimates)**
Using `Counter`, we compute raw counts only on the **training data**:

$$
\begin{aligned}
C(w) &\quad \text{(unigram count)} \\
C(w_{i-1}, w_i) &\quad \text{(bigram count)} \\
C(w_{i-2}, w_{i-1}, w_i) &\quad \text{(trigram count)}
\end{aligned}
$$

Let:
- $ N $ = total number of tokens in training
- $ V $ = vocabulary size = number of unique unigrams

#### 4. **Interpolated Trigram Model with Laplace Smoothing**

The core probability function computes:

$$
P(w_i \mid w_{i-2}, w_{i-1}) = \lambda_3 P_3(w_i \mid w_{i-2}, w_{i-1}) + \lambda_2 P_2(w_i \mid w_{i-1}) + \lambda_1 P_1(w_i)
$$

where $\lambda_1 + \lambda_2 + \lambda_3 = 1$ and here:  
**$\lambda_1 = 0.1$, $\lambda_2 = 0.3$, $\lambda_3 = 0.6$**

Each component uses **Add-α (Laplace) smoothing** with $\alpha = 1$:

$$
\begin{aligned}
P_3(w_i \mid w_{i-2},w_{i-1}) &=
\begin{cases}
\dfrac{C(w_{i-2},w_{i-1},w_i) + 1}{C(w_{i-2},w_{i-1}) + V} & \text{if } C(w_{i-2},w_{i-1}) > 0 \\
0 & \text{otherwise (falls back to lower order)}
\end{cases} \\[10pt]
\\
P_2(w_i \mid w_{i-1}) &= \dfrac{C(w_{i-1},w_i) + 1}{C(w_{i-1}) + V} \\[10pt]
\\
P_1(w_i) &= \dfrac{C(w_i) + 1}{N + V}
\end{aligned}
$$

This ensures **zero probability never occurs** and provides smooth fallback from trigram → bigram → unigram.

#### 5. **Intrinsic Evaluation: Perplexity**

Perplexity is the standard measure of predictive performance:

$$
\text{PP}(M) = \exp\left( -\frac{1}{T} \sum_{i=1}^{T} \log P(w_i \mid w_{i-2}, w_{i-1}) \right) = P(w_1...w_T)^{-1/T}
$$

where $T$ is the number of predicted tokens in the test set.

- Lower perplexity = better model
- Two `<s>` symbols are prepended and one `</s>` appended to handle sentence boundaries
- Logarithms are used for numerical stability

**Interpretation**: A perplexity of ~500–700 is **excellent** for a classical n-gram model on raw real-world Amharic text.

#### 6. **Extrinsic Evaluation: Text Generation via Backoff Sampling**

The `generate_sentence()` function performs **ancestral sampling** using a **trigram → bigram → unigram backoff strategy**:

1. Start with a frequent bigram sampled proportional to its training count:
   $$
   (w_1, w_2) \sim P_{ML}(w_1, w_2)
   $$

2. For each subsequent word $w_t$, attempt to sample from:
   - Highest order available:  
     $ w_t \sim P(w_t \mid w_{t-2}, w_{t-1}) $ (trigram)  
     → if none exist → fall back to bigram → then unigram

3. Stop when an end-of-sentence punctuation (።, ፧, !, ?) is generated or max length reached.

This produces **locally coherent, grammatically natural Amharic sentences** that reflect patterns learned from the real corpus.


````python
#  CLEAN n-gram generation (flat tuples of strings)
def generate_ngrams(tokens, n):
    """Returns list of tuples like ('w1', 'w2', ...) – never nested"""
    return [tuple(tokens[i:i+n]) for i in range(len(tokens)-n+1)]
print("tokens size:" ,len(tokens))
# Train / Test split (90% / 10%)
split = int(0.9 * len(tokens))
train_tokens = tokens[:split]
test_tokens  = tokens[split:]

print(f"Train tokens : {len(train_tokens):,}")
print(f"Test  tokens : {len(test_tokens):,}")

# Build frequency counters on TRAIN data only
from collections import Counter

train_unigram  = Counter(generate_ngrams(train_tokens, 1))
train_bigram   = Counter(generate_ngrams(train_tokens, 2))
train_trigram  = Counter(generate_ngrams(train_tokens, 3))

V = len(train_unigram)                     # vocabulary size
N = len(train_tokens)                      # total tokens in training

# Interpolated Trigram Model with Laplace smoothing
alpha = 1.0
l1, l2, l3 = 0.1, 0.3, 0.6                 # interpolation weights

def interpolated_prob(w1, w2, w3):
    # w1, w2, w3 are strings (not tuples!)
    c3 = train_trigram.get((w1, w2, w3), 0)
    c2 = train_bigram.get((w1, w2), 0)
    c2_next = train_bigram.get((w2, w3), 0)
    c1_prev = train_unigram.get((w2,), 0)
    c1 = train_unigram.get((w3,), 0)

    # Trigram probability (Laplace)
    p3 = (c3 + alpha) / (c2 + alpha * V) if c2 > 0 else 0

    # Bigram probability
    p2 = (c2_next + alpha) / (c1_prev + alpha * V) if c1_prev > 0 else 0

    # Unigram probability
    p1 = (c1 + alpha) / (N + alpha * V)

    return l3 * p3 + l2 * p2 + l1 * p1

# INTRINSIC EVALUATION – Perplexity
import math

def perplexity(tokens_to_evaluate):
    words = ["<s>", "<s>"] + tokens_to_evaluate + ["</s>"]
    log_prob = 0.0
    count = 0
    for i in range(2, len(words)):
        p = interpolated_prob(words[i-2], words[i-1], words[i])
        if p == 0: p = 1e-12
        log_prob += math.log(p)
        count += 1
    return math.exp(-log_prob / count)

ppl = perplexity(test_tokens)
print(f"\n=== INTRINSIC EVALUATION ===")
print(f"Perplexity on held-out test set ({len(test_tokens):,} tokens) = {ppl:.2f}")

# EXTRINSIC EVALUATION – Text Generation (Hybrid back-off)
import random

END_PUNCT = {"።", "፧", "!", "?"}

def weighted_choice(choices, weights):
    return random.choices(choices, weights=weights, k=1)[0]

def generate_sentence(max_len=25):
    # Start with a frequent bigram
    start_bigram = weighted_choice(
        list(train_bigram.keys()),
        [train_bigram[b] for b in train_bigram.keys()]
    )
    sent = list(start_bigram)                 # two starting words

    for _ in range(max_len):
        w1, w2 = sent[-2], sent[-1]

        # 1) Trigram
        candidates_tri = [(w1,w2,w3) for (w1_,w2_,w3) in train_trigram.keys() if w1_==w1 and w2_==w2]
        if candidates_tri:
            next_w = weighted_choice(
                [w3 for (_,_,w3) in candidates_tri],
                [train_trigram[t] for t in candidates_tri]
            )
            sent.append(next_w)
            if next_w in END_PUNCT:
                break
            continue

        # 2) Bigram back-off
        candidates_bi = [(w2,w3) for (w2_,w3) in train_bigram.keys() if w2_==w2]
        if candidates_bi:
            next_w = weighted_choice(
                [w3 for (_,w3) in candidates_bi],
                [train_bigram[b] for b in candidates_bi]
            )
            sent.append(next_w)
            if next_w in END_PUNCT:
                break
            continue

        # 3) Unigram back-off
        next_w = weighted_choice(
            [w[0] for w in train_unigram.keys()],
            list(train_unigram.values())
        )
        sent.append(next_w)
        if next_w in END_PUNCT:
            break

    return " ".join(sent)


print("\n=== EXTRINSIC EVALUATION – Generated Sentences ===")
print("Rate fluency & grammaticality (0–5):")
for i in range(8):
    s = generate_sentence(max_len=30)
    print(f"{i+1:2}. {s}")

# Save 50 generated sentences to Drive
with open("/content/drive/MyDrive/NLP/amharic_ngram_generated.txt", "w", encoding="utf-8") as f:
    for _ in range(50):
        f.write(generate_sentence(max_len=35) + "\n\n")

print("\n50 generated sentences saved to Drive!")
print("\nDone – both intrinsic (perplexity) and extrinsic (generation) evaluation complete!")
````
